```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
G= 6.67430*(10**(-8))
c= 2.99792458*10**10

k= 5.38*(10**9)
gamma= 1.667
```

```python
def f(x):
    
    r=x[0]
    m=x[1]
    P=x[2]
    
    rho=(P/k)**(1/gamma)
    
    
    dr_dr=1
    dm_dr=4*np.pi*(r**2)*rho
    dP_dr=-((G*m*rho)/(r**2))*(1+(P/(rho*(c**2))))*(1+((4*np.pi*P*(r**3))/(m*(c**2))))*((1-((2*m*G)/(r*(c**2))))**(-1))
    
    return np.array([dr_dr,dm_dr,dP_dr])
```

```python
def ns_solve(rho_c):
    
    h=100
    P_c=k*(rho_c)**(gamma)
    
    X=np.zeros([3,80000])
    
    r0 = 10
    m0 = (4 / 3) * np.pi * (r0**3) * rho_c
    
    
    
    X[:,0]=np.array([r0,m0,P_c])
    
    #r=r0
    
    
    radius=[]
    mass=[]
    pressure=[]
    density=[]
    
    for i in range(1, 80000):
        k1 = h * f(X[:, i-1])
        k2 = h * f(X[:, i-1] + 0.5 * k1)
        k3 = h * f(X[:, i-1] + 0.5 * k2)
        k4 = h * f(X[:, i-1] + k3)
        
        
        X[:, i] = X[:, i-1] + (k1 + 2 * k2 + 2 * k3 + k4) / 6

        
    #for i in range(1, 80000):
        #k1 = h * f(r, X[:, i-1])  # k1 = h * f(r, x)
        #k2 = h * f(r + 0.5 * h, X[:, i-1] + 0.5 * k1)  # k2 = h * f(r + h/2, x + k1/2)
        #k3 = h * f(r + 0.5 * h, X[:, i-1] + 0.5 * k2)  # k3 = h * f(r + h/2, x + k2/2)
        #k4 = h * f(r + h, X[:, i-1] + k3)  # k4 = h * f(r + h, x + k3)
        
        #X[:, i] = X[:, i-1] + (k1 + 2 * k2 + 2 * k3 + k4) / 6  # Update step
    
    
        
        if((X[2,i]/P_c)<1e-10):
            break
        
        #r += h  # Update the independent variable
        radius.append(X[0,i-1])
        mass.append(X[1,i-1])
        pressure.append(X[2,i-1])
        density.append((X[2,i-1]/k)**(1/gamma))
        
        
        #radius.append(r)
        #mass.append(X[0, i-1])
        #pressure.append(X[1, i-1])
        #density.append((X[1, i-1] / k) ** (1 / gamma))
        
    return X[:,i-1],radius,mass,pressure,density


        
    
    
```

```python
G=ns_solve(1*10**15)
```

```python
#print(G[4])
```

```python
print("The central pressure we consider is",k*(10**15)**(gamma),"dyn/cm^2")
print("The surface radius at this central pressure is",G[0][0],"cm")
print("The total mass at this central pressure is",G[0][1],"gm")
```

```python
plt.plot(G[1],G[2],"g")
plt.title("Enclosed mass vs radial distance graph",color="blue",size=17)
plt.xlabel("r(cm)", fontsize=14, fontweight='bold')
plt.ylabel("$m(gm)$", fontsize=14, fontweight='bold')
plt.show()
```

```python
plt.plot(G[1],G[3],"k")
plt.title("Enclosed pressure vs radial distance graph",color="blue",size=17)
plt.xlabel("r(cm)", fontsize=14, fontweight='bold')
plt.ylabel("p($dyn/cm^2$)", fontsize=14, fontweight='bold')
plt.show()
```
